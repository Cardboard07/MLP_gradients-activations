```python
import pandas as pd
import torch as t
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
data=pd.read_csv(r"C:\Users\hrish\Projects\Deep_learning_projects\Trigram\NationalNames.csv")
```

```python
names=data['Name'].unique()
```

```python
random.seed(42)
names=list(names)
random.shuffle(names)
```

```python
names=[n.lower() for n in names]
```

```python
alpha=list(set(''.join(names)))
alpha.sort()
```

```python
con={a:i+1 for i,a in enumerate(alpha)}
con['.']=0
```

```python
X=[]
y=[]
for nm in names:
    ct=[0,0,0]
    nm=nm+'.'
    for c in nm:
        X.append(ct)
        y.append(con[c])
        ct=ct[1:]+[con[c]]
```

```python
g = t.Generator().manual_seed(2147483647)
C = t.randn((27, 10), generator=g)
W1 = t.randn((30, 500), generator=g)*0
b1 = t.randn(500, generator=g)*0
W2= t.randn((500, 500), generator=g)*0
b2= t.randn(500, generator=g)*0
W3 = t.randn((500, 27), generator=g)*0
b3 = t.randn(27, generator=g)*0
parameters = [C, W1, b1, W2, b2, W3, b3]
```

```python
for p in parameters:
    p.requires_grad=True
```

```python
Xtr=t.tensor(X[:500001])
ytr=t.tensor(y[:500001])
Xts=t.tensor(X[500001:])
yts=t.tensor(y[500001:])
```

```python
C[Xtr[:5]].view(-1,30)
```

```python
lr=1
```

```python
lossi=[]
for i in range(40000):
    ix=t.randint(0,Xtr.shape[0],(1000,),generator=g)
    h1=C[Xtr[ix]].view(-1,30)@W1+b1
    h1=t.tanh(h1)
    h2=h1@W2+b2
    h2=t.tanh(h2)
    logits=h2@W3+b3
    loss=loss=t.nn.functional.cross_entropy(logits,ytr[ix])
    for p in parameters:
        p.grad=None
    loss.backward()
    if(i>20000):
        lr=0.1
    for p in parameters:
        p.data-=lr*p.grad
    lossi.append(loss)
```

```python
layers=[t.nn.Linear(30,500),t.nn.BatchNorm1d(500),t.nn.Tanh(),t.nn.Linear(500, 500),t.nn.BatchNorm1d(500),t.nn.Tanh(),t.nn.Linear(500, 27)]
parameters = [C] + [p for layer in layers for p in layer.parameters()]
for p in parameters:
  p.requires_grad = True
```

```python
lossi=[]
for i in range(40000):
    ix=t.randint(0,Xtr.shape[0],(1000,),generator=g)
    emb=C[Xtr[ix]].view(-1,30)
    res=emb
    for l in layers:
        res=l(res)
    loss=t.nn.functional.cross_entropy(res,ytr[ix])
    for p in parameters:
        p.grad=None
    loss.backward()
    if(i>20000):
        lr=0.1
    for p in parameters:
        p.data-=lr*p.grad
    lossi.append(loss)
```

```python
emb=C[Xtr]
res=emb
for l in layers:
    res=l(res)
loss=t.nn.functional.cross_entropy(res,ytr[ix])
```

```python
with torch.no_grad():
    k=layers[1].weight/((layers[1].running_var+layers[1].eps)**0.5)
    layers[0].weight=layers[0].weight*k.view(-1,1)
    layers[0].bias=k*(layers[0].bias-layers[1].running_mean)+layers[1].bias
```

```python
with torch.no_grad():
    k=layers[4].weight/((layers[4].running_var+layers[4].eps)**0.5)
    layers[3].weight=layers[3].weight*k.view(-1,1)
    layers[3].bias=k*(layers[3].bias-layers[4].running_mean)+layers[4].bias
```

```python
new_layers=[layers[0]+layers[2]+layers[3]+layers[5]+layers[-1]]
```

```python
emb=C[Xtr]
res=emb
for l in new_layers:
    res=l(res)
loss=t.nn.functional.cross_entropy(res,ytr[ix])
```
